```python
import pandas as pd
import sqlite3
from pathlib import Path
```

```python
base_path = Path(r"C:\Users\Rubah\Downloads\Final_Project")
processed_data_path_core = base_path / "Data/processed/core"
processed_data_path_lookup = base_path / "Data/processed/lookup"
db_path = base_path / "Database/nabahan.db"
```

```python
db_path.parent.mkdir(parents=True, exist_ok=True)
```

```python
files = {
    'future_projects': processed_data_path_core / "future_projects_clean.csv",
    'tenders_full_details': processed_data_path_core / "nabahan_full_details_clean.csv",
    'government_entity': processed_data_path_lookup / "government_entity_clean.csv",
    'primary_activity': processed_data_path_lookup / "primary_activity_clean.csv",
    'secondary_activity': processed_data_path_lookup / "secondary_activity_clean.csv",
    'regions': processed_data_path_lookup / "regions_clean.csv",
    'tender_doc_price_range': processed_data_path_lookup / "tender_doc_price_range_clean.csv",
    'tender_publication_date': processed_data_path_lookup / "tender_publication_date_clean.csv",
    'tender_statuses': processed_data_path_lookup / "tender_statuses_clean.csv",
    'tender_types': processed_data_path_lookup / "tender_types_clean.csv"
}
```

```python
conn = sqlite3.connect(db_path)
```

```python
for table_name, file_path in files.items():
    df = pd.read_csv(file_path)
    df.to_sql(table_name, conn, if_exists='replace', index=False)
    print(f"âœ… Created table '{table_name}' from {file_path.name}")
```

```python
try:
    sample = pd.read_sql("SELECT * FROM tenders_full_details LIMIT 3", conn)
    print("\n--- Sample from 'tenders_full_details' ---")
    print(sample)

    row_count = pd.read_sql("SELECT COUNT(*) AS total_rows FROM future_projects", conn)
    print(f"\nTotal rows in 'future_projects': {row_count.iloc[0]['total_rows']}")

finally:
    conn.close()
    print("\nðŸ”’ Database connection closed safely.")
```
